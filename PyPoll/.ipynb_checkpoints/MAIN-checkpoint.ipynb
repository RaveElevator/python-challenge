````python
#set variables
winner = 0
candidates = []
finallist = []
firstlist = []

#Set a file path
read_file = os.path.join("election_data.csv")
````

````python
candidates = []
with open(read_file, "r", newline = "") as pypollfile:
    
    #set csv reader
    csvreader = csv.reader(pypollfile, delimiter = ',')
    
    for row in csvreader:
        #put the colums in their own list, making it easy for us
        candidates.append(row[2])
candidates.pop(0)
candidates
````

````python
#  * The total number of votes cast
len(candidates)
````

````python
#  * A complete list of candidates who received votes

countlist = [[x,candidates.count(x)] for x in set(candidates)]
countlist
````

````python
#  * The percentage of votes each candidate won

Correy_percent = countlist[0][1] / len(candidates) * 100
Li_percent = countlist[1][1] / len(candidates) * 100
OTooley_percent = countlist[2][1] / len(candidates) * 100
Kahn_percent = countlist[3][1] / len(candidates) * 100

print(str(Correy_percent))
print(str(Li_percent))
print(str(OTooley_percent))
print(str(Kahn_percent))
````

````python
# The Winner is:  The winner of the election based on popular vote.

votecount = 0
for item in countlist:
    vote = item[1]
    print(item)
    if vote > votecount:
        votecount = vote
        winner = item[0]

winner
````
